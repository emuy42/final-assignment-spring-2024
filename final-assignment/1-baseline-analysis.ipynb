```python
## import library
import pandas as pd
```

```python
## set display code
pd.options.display.float_format = '{:,.0f}'.format
```

```python
## import raw dataset 
raw_df = pd.read_excel("college-encampment-database.xlsx", sheet_name='6.18_workbook')
raw_df
```

```python
## duplicate raw data
df_analysis = raw_df
```

```python
df_analysis.info()
```

## convert all strings to numbers

```python
## convert all other metrics to int
df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]]= df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]].astype("float")
```

# Step 1: Take a baseline of the whole data set

## find top states for each

```python
## find top states
(df_analysis.value_counts("state", normalize = True)*100).to_frame("State breakdown")
```

## public v. private split

```python
## find public v. private split for each - success
(df_analysis.value_counts("school_type", normalize = True)*100).to_frame("Public v. Private")
```

## mean, median, max and min endowment size

```python
## find mean, median, max, min endowment size
df_analysis["endowment_size"].agg(["mean", "median", "max", "min"]).to_frame("Endowment size")
```

## mean, median, max and min tuition

```python
## find mean, median, max, min tuition - success
df_analysis[["undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private"]].agg(["mean", "median", "max", "min"])
```

## mean, median, max and min encampment length

```python
## find mean, median, max, min encampment length - success
df_analysis["length_encampment"].agg(["mean", "median", "max", "min"]).to_frame("Encampment length")
```

## mean, median, max and min encampment size

```python
## find mean, median, max, min encampment size
df_analysis["size_encampment"].agg(["mean", "median", "max", "min"]).to_frame("Encampment size")
```

## mean, median, max and min arrested total

```python
## find mean, median, max, min arrested total
df_analysis["arrested_total"].agg(["mean", "median", "max", "min"]).to_frame("Total arrests")
```

## mean, median, max and min undergrad population

```python
## find mean, median, max, min arrested total
df_analysis["undergrad_population"].agg(["mean", "median", "max", "min"]).to_frame("Undergrad population")
```

## share of schools that had counter protesters

```python
## find share of schools that had counter protestors
(df_analysis.value_counts("counterprotesters", normalize = True)*100).to_frame("Had counterprotesters?")
```

## share of schools that are in states with anti-BDS laws

```python
## find share of schools that have BDS laws
(df_analysis.value_counts("anti_bds_state", normalize = True)*100).to_frame("Anti BDS law?")
```

## share of schools with arrested students

```python
## find share of schools with arrested students
(df_analysis.value_counts("university_response", normalize = True)*100).to_frame("School response")
```
