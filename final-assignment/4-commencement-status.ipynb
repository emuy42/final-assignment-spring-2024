```python
## import library
import pandas as pd
```

```python
## set display code
pd.options.display.float_format = '{:,.0f}'.format
```

```python
## import raw dataset 
raw_df = pd.read_excel("college-encampment-database.xlsx", sheet_name='6.18_workbook')
raw_df
```

```python
## duplicate raw data
df_analysis = raw_df
```

```python
df_analysis.info()
```

## convert all strings to numbers

```python
## convert all other metrics to int
df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]]= df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]].astype("float")
```

# Step 4: Are there discernible differences among schools that cancelled commencement?

## share of schools that canceled commencement status

```python
## value count of schools' commencement decisions
df_analysis.value_counts("commencement_canceled").to_frame("Commencement status")
```

```python
## share of schools that called the police
(df_analysis.value_counts("commencement_canceled", normalize = True)*100).to_frame("Commencement status")
```

## create DFs for analysis

```python
## create a separate DF to look at schools that have canceled commencement
df_canceled = df_analysis[["demand_scope","start_encampment","end_encampment","encampment_ongoing","size_encampment","length_encampment","counterprotesters","commencement_canceled","university_response","arrested_total","negotiated","negotiation_date","school_type","endowment_size","undergrad_population","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","anti_bds_state","state"]].query("university_response == 'arrests'")
df_canceled.sample(10)
```

```python
## create a separate DF to look at schools that did NOT cancel commencement
df_commencement = df_analysis[["demands","start_encampment","end_encampment","encampment_ongoing","size_encampment","length_encampment","counterprotesters","commencement_canceled","university_response","arrested_total","negotiated","negotiation_date","school_type","endowment_size","undergrad_population","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","anti_bds_state","state"]].query("university_response == 'arrests'")
df_commencement.sample(10)
```

## find top states for each

```python
## find top states for each - canceled
df_canceled["state"].value_counts(ascending = False).to_frame("Total schools - commencement canceled")
```

```python
## find top states for each - commencement
df_commencement["state"].value_counts(ascending = False).to_frame("Total schools - commencement happened")
```

## public v. private split

```python
## find public v. private split for each - commencement canceled
(df_canceled.value_counts("school_type", normalize = True)*100).to_frame("Public v. Private")
```

```python
## find public v. private split for each - commencement happened
(df_commencement.value_counts("school_type", normalize = True)*100).to_frame("Public v. Private")
```

```python
## find mean, median, max, min endowment size - commencement canceled
df_canceled["endowment_size"].agg(["mean", "median", "max", "min"]).to_frame("Endowment size - ccommencement canceled")
```

```python
## find mean, median, max, min endowment size - commencement happened
df_commencement["endowment_size"].agg(["mean", "median", "max", "min"]).to_frame("Endowment size - commencement happened")
```
