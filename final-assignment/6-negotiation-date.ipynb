```python
## import library
import pandas as pd
```

```python
## set display code
pd.options.display.float_format = '{:,.0f}'.format
```

```python
## import raw dataset 
raw_df = pd.read_excel("college-encampment-database.xlsx", sheet_name='6.18_workbook')
raw_df
```

```python
## duplicate raw data
df_analysis = raw_df
```

```python
df_analysis.info()
```

## convert all strings to numbers

```python
## convert all other metrics to int
df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]]= df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]].astype("float")
```

# Step 5: Is there a date after which negotiation is more or less likely? Or a specific encampment duration?

## create DF

```python
## create DF
df_negotiation = df_analysis[["demand_scope","start_encampment","end_encampment","encampment_ongoing","size_encampment","length_encampment","counterprotesters","commencement_canceled","university_response","arrested_total","negotiated","negotiation_date","school_type","endowment_size","undergrad_population","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","anti_bds_state","state"]].query("negotiated == 'yes' or (negotiated == 'success')")
df_negotiation.sample(10)
```

## find mean, media, max, min for negotiation dates

```python
## import datetime
import datetime
```

```python
## convert negotiation date into datetime for analysis
df_negotiation["negotiation_date"] = pd.to_datetime(df_negotiation["negotiation_date"])
```

```python
## find mean, median, max, and min for negotiation date
df_negotiation["negotiation_date"].agg(["mean", "median", "max", "min"]).to_frame("Key stats - negotiation date")
```

## find mean, median, max, min for encampment length

```python
## find mean, median, max, and min for encampment length
df_negotiation["length_encampment"].agg(["mean", "median", "max", "min"]).to_frame("Key stats - encampment length")
```

## demand scope

```python
(df_negotiation.value_counts("demand_scope", normalize = True)*100).to_frame("Demand scope")
```
