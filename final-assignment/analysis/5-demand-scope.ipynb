```python
## import library
import pandas as pd
```

```python
## set display code
pd.options.display.float_format = '{:,.0f}'.format
```

```python
## import raw dataset 
raw_df = pd.read_excel("college-encampment-database.xlsx", sheet_name='6.18_workbook')
raw_df
```

```python
## duplicate raw data
df_analysis = raw_df
```

```python
df_analysis.info()
```

## convert all strings to numbers

```python
## convert all other metrics to int
df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]]= df_analysis[["endowment_size","size_encampment","length_encampment","arrested_total","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","undergrad_population"]].astype("float")
```

# Step 5: Are there discernible differences among schools that have different scopes of demand?

## share of schools by demand 

```python
## value count of schools' commencement decisions
df_analysis.value_counts("demand_scope").to_frame("Demand scope")
```

```python
## share of schools that called the police
(df_analysis.value_counts("demand_scope", normalize = True)*100).to_frame("Demand scope")
```

## create DFs for analysis

```python
## create a separate DF to look at schools that have a broad scope
df_broad = df_analysis[["demand_scope","start_encampment","end_encampment","encampment_ongoing","size_encampment","length_encampment","counterprotesters","commencement_canceled","university_response","arrested_total","negotiated","negotiation_date","school_type","endowment_size","undergrad_population","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","anti_bds_state","state"]].query("demand_scope == 'broad'")
df_broad.sample(10)
```

```python
## create a separate DF to look at schools that have a standard scope
df_standard = df_analysis[["demand_scope","start_encampment","end_encampment","encampment_ongoing","size_encampment","length_encampment","counterprotesters","commencement_canceled","university_response","arrested_total","negotiated","negotiation_date","school_type","endowment_size","undergrad_population","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","anti_bds_state","state"]].query("demand_scope == 'standard'")
df_standard.sample(10)
```

```python
## create a separate DF to look at schools that have a narrow scope
df_narrow = df_analysis[["demand_scope","start_encampment","end_encampment","encampment_ongoing","size_encampment","length_encampment","counterprotesters","commencement_canceled","university_response","arrested_total","negotiated","negotiation_date","school_type","endowment_size","undergrad_population","undergrad_tuition_instate","undergrad_tuition_outstate","undergrad_tuition_private","anti_bds_state","state"]].query("demand_scope == 'narrow'")
df_narrow.sample(10)
```

## find top states for each

```python
## find top states for each - broad
df_broad["state"].value_counts(ascending = False).to_frame("Total schools - broad")
```

```python
## find top states for each - standard
df_standard["state"].value_counts(ascending = False).to_frame("Total schools - standard")
```

```python
## find top states for each - narrow
df_narrow["state"].value_counts(ascending = False).to_frame("Total schools - narrow")
```
